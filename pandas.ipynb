#### Задание 1. Соберите всю информацию о студентах в одну таблицу df. В получившейся таблице должна быть информация и оценки всех студентов из всех групп. Напечатайте несколько строк таблицы для демонстрации результата.

```python
df_students = pd.read_csv('Students_info_0.csv') #считываем первый файл с студентами

df_marks = pd.read_csv('Students_marks_0.csv') #считываем первый файл с оценками

for i in range(1,10):
  df_students = pd.concat([df_students, pd.read_csv(f'Students_info_{i}.csv')], ignore_index=True) # соединяем дата фреймы вертикально пересчитывая индекс студента
  df_marks = pd.concat([df_marks, pd.read_csv(f'Students_marks_{i}.csv')], ignore_index=True)      # аналогично поступаем с дата фремами оценок

df = df_students.join(df_marks, lsuffix='_student', rsuffix='_mark')
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
def GetGrade(average_mark):
    if average_mark >= 90:
      return 'A'
    if 80 <= average_mark < 90:
      return 'B'
    if 70 <= average_mark < 80:
      return 'C'
    if 60 <= average_mark < 70:
      return 'D'
    if 50 <= average_mark < 60:
      return 'E'
    return 'F'
```

```python
df.info() #посмотри все ли хорошо
```

```python
df['Grade'] = np.mean(df[['math score', 'reading score', 'writing score']], axis = 1) #находим средние по строкам вдоль оценок по математике, чтению и письму
```

```python
df['Grade'] = df.apply(lambda x: GetGrade(x['Grade']), axis=1) #применяем функцию GetGrade к столбцу 'Grade'
```
